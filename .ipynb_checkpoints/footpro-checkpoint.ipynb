```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error
import numpy as np

# Load data
data = pd.read_csv('filtered_data.csv')

# Preprocess the data
data['match_outcome'] = data['result'].map({'W': 1, 'L': 0, 'D': 2})

# Define the features for prediction
features = ['gf', 'ga', 'xg', 'xga', 'poss', 'sh', 'dist', 'fk', 'pk', 'team', 'opponent']

# Define the targets for regression (xg, xga, poss for team and opponent)
X = data[features]
y_team_xg = data['xg']
y_opponent_xg = data['xga']
y_team_poss = data['poss']

# Step 1: Create preprocessor for categorical and numerical features
categorical_columns = ['team', 'opponent']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), ['gf', 'ga', 'sh', 'dist', 'fk', 'pk']),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns)
    ])

# Step 2: Build the pipeline for the regression model (xg, xga, poss prediction for team)
regression_pipeline_team_xg = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])

regression_pipeline_opponent_xg = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])

regression_pipeline_team_poss = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])

# Step 3: Train-test split for regression targets
X_train, X_test, y_train_team_xg, y_test_team_xg = train_test_split(X, y_team_xg, test_size=0.2, random_state=42)
y_train_opponent_xg, y_test_opponent_xg = train_test_split(y_opponent_xg, test_size=0.2, random_state=42)
y_train_team_poss, y_test_team_poss = train_test_split(y_team_poss, test_size=0.2, random_state=42)

# Step 4: Fit the regression models
regression_pipeline_team_xg.fit(X_train, y_train_team_xg)
regression_pipeline_opponent_xg.fit(X_train, y_train_opponent_xg)
regression_pipeline_team_poss.fit(X_train, y_train_team_poss)

# Step 5: Evaluate the models for performance
y_pred_team_xg = regression_pipeline_team_xg.predict(X_test)
y_pred_opponent_xg = regression_pipeline_opponent_xg.predict(X_test)
y_pred_team_poss = regression_pipeline_team_poss.predict(X_test)

print("Mean Absolute Error for Team xG:", mean_absolute_error(y_test_team_xg, y_pred_team_xg))
print("Mean Absolute Error for Opponent xG:", mean_absolute_error(y_test_opponent_xg, y_pred_opponent_xg))
print("Mean Absolute Error for Team Possession:", mean_absolute_error(y_test_team_poss, y_pred_team_poss))

# Step 6: Predict Expected Goals (xg), xGA, and Possession for selected teams
def predict_match_stats(team, opponent):
    sample_data = {
        'gf': [0], 'ga': [0], 'xg': [0.0], 'xga': [0.0], 'poss': [50], 'sh': [0],
        'dist': [0.0], 'fk': [0], 'pk': [0], 'team': [team], 'opponent': [opponent]
    }
    
    sample_df = pd.DataFrame(sample_data)
    
    # Preprocessed data
    predicted_team_xg = regression_pipeline_team_xg.predict(sample_df)
    predicted_opponent_xg = regression_pipeline_opponent_xg.predict(sample_df)
    predicted_team_poss = regression_pipeline_team_poss.predict(sample_df)

    return {
        'Predicted Team xG': predicted_team_xg[0],
        'Predicted Opponent xGA': predicted_opponent_xg[0],
        'Predicted Team Possession': predicted_team_poss[0]
    }
    # Example test with two teams
team = "Manchester City"
opponent = "Tottenham"
predictions = predict_match_stats(team, opponent)
print("\nPredictions for {} vs {}:".format(team, opponent))
print(predictions)
```
